```python
import csv
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
# Load dataset and select all columns that we will use.
DATA_PATH = r"../data/cancer_patient_data_sets.csv"
raw_dataframe = pd.read_csv(DATA_PATH)

raw_dataframe.head(10)
```

```python
cleaned_df = raw_dataframe.loc[:, raw_dataframe.columns != "patient_id"]

# Train and test split
n_sample = cleaned_df.shape[0]
np.random.seed(0)

mask = np.random.rand(n_sample) < 0.8

training_set = cleaned_df[mask]
testing_set = cleaned_df[~mask]

training_X = training_set.loc[:, training_set.columns != "level"]
training_y = np.array(training_set[["level"]]).reshape((training_set.shape[0], )).T

print(f"Training set contains {training_set.shape[0]} entries;\nTesting set contains {testing_set.shape[0]} entries.")
```

```python
# Train a random forest based on these data
nn = RandomForestClassifier()
nn = nn.fit(training_X, training_y)
```

```python
# Testing the model
testing_X = testing_set.loc[:, testing_set.columns != "level"]
testing_y = testing_set[["level"]]

nn.score(testing_X, testing_y)
```

```python
for idx, row in testing_X.iterrows():
    y_hat = nn.predict(row.to_numpy().reshape((-1, 23)))  # 23 features
    y_true = testing_y.loc[idx]["level"]
    print(f"predict: {y_hat[0]}, truth: {y_true}\n")
```
